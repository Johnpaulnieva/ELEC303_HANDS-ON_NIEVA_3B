<h2>6.) CHECK FOR NULL VALUES</h2>

```python
df.isnull().sum()
```

```python
df['Location Description'] = df['Location Description'].fillna(df['Location Description'].mode()[0])
df['Location'] = df['Location'].fillna(df['Location'].mode()[0])
df['Community Area'] = df['Community Area'].fillna(df['Community Area'].mean())
df['X Coordinate'] = df['X Coordinate'].fillna(df['X Coordinate'].mean())
df['Y Coordinate'] = df['Y Coordinate'].fillna(df['Y Coordinate'].mean())
df['Latitude'] = df['Latitude'].fillna(df['Latitude'].mean())
df['Longitude'] = df['Longitude'].fillna(df['Longitude'].mean())
```

```python
df.isnull().sum()
```

<h2>8.) CHECK COLUMN DATA TYPES</h2>

```python
df.info()
```

```python
df.dtypes
```

```python
df['Case Number'] = df ['Case Number'].astype('category')
df['Date'] = df ['Date'].astype('category')
df['Block'] = df ['Block'].astype('category')
df['IUCR'] = df ['IUCR'].astype('category')
df['Primary Type'] = df ['Primary Type'].astype('category')
df['Description'] = df ['Description'].astype('category')
df['Location Description'] = df ['Location Description'].astype('category')
df['FBI Code'] = df ['FBI Code'].astype('category')
df['Updated On'] = df ['Updated On'].astype('category')
df['Location'] = df ['Location'].astype('category')
```

```python
df.dtypes
```

<h2>CLEARING ANY INCONSISTENCIES WITH THE DATA SET</h2>
<hr>

```python
df['Date'] = df['Date'].str.strip() #to remove any extra white space
df['Date'] = df['Date'].astype('str') #to convert into a string
df['Date'] = df['Date'].str.replace('/', '-')
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst = True, errors = 'coerce')
```

```python
df.dtypes
```

<h2>EXTRACTING DATE INFORMATION USING PANDAS DATE TIME</h2>
<hr>

```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayOfWeek'] = df['Date'].dt.dayofweek #Monday=0, Sunday=6
```

```python
df.isnull().sum()
```

```python
for i in df['Location Description'].unique():
    print(i)
```

```python
street_df = df[df['Location Description'] == 'STREET']
```

```python
yearly_streetdata = street_df.groupby(['Year', 'Month']).size()

yearly_streetdata = yearly_streetdata.unstack(level=0)

yearly_streetdata.plot(
    marker='o', 
    figsize=(10, 5),
    title='yearly Crimes in Street by Year (2024-2025)'
)
plt.xticks(range(1, 13), ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.legend(title='Yearly Crimes Comparison', bbox_to_anchor = (1.05,1), loc='upper left')
plt.grid(alpha=0.3)
plt.show()
```

```python
monthly_arrests = df.groupby(['Year','Month'])['Arrest'].mean().reset_index()
monthly_arrests['Arrest Rate'] = monthly_arrests['Arrest'] * 100
monthly_arrests_pivot = monthly_arrests.pivot(index='Month', columns='Year', values='Arrest Rate')

monthly_arrests_pivot.plot(marker='o', figsize=(12,6))
plt.title("Monthly Arrest Rates in Chicago by Year")
plt.xticks(range(1,13), ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.ylabel("Arrest Rate (%)")
plt.xlabel("Month")
plt.legend(title="Year", bbox_to_anchor=(1.05,1), loc="upper left")
plt.grid(alpha=0.3)
plt.show()
```

```python
# Ensure Date is datetime
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

# Group crimes per month & year
monthly_crime = df.groupby(['Year','Month']).size().reset_index(name='Count')
monthly_crime_pivot = monthly_crime.pivot(index='Month', columns='Year', values='Count')

# Plot
monthly_crime_pivot.plot(marker='o', figsize=(12,6))
plt.title("Monthly Crime Trends in Chicago by Year")
plt.xticks(range(1,13), ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.ylabel("Number of Crimes")
plt.xlabel("Month")
plt.legend(title="Year", bbox_to_anchor=(1.05,1), loc="upper left")
plt.grid(alpha=0.3)
plt.show()
```

```python
yearly_crime = df.groupby('Year').size().reset_index(name='count')
yearly_crime.plot(x='Year', y='count', kind='bar', figsize=(10,6), legend=False)
plt.title("Crimes per Year")
plt.ylabel("Number of Crimes")
plt.xlabel("Year")
plt.show()

```

```python
locations = df.groupby('Location Description').size().sort_values(ascending=False).head(10)
locations.plot(kind='barh', figsize=(10,6))
plt.title("Top 10 Locations of Crimes")
plt.xlabel("Number of Crimes")
plt.ylabel("Location")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x="Month", data=df, palette="coolwarm")
plt.title("Crimes by Month")
plt.xticks(range(12), ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'])
plt.show()
```

```python
df['Hour'] = df['Date'].dt.hour

plt.figure(figsize=(12,6))
sns.countplot(x="Hour", data=df, palette="plasma")
plt.title("Crimes by Hour of the Day")
plt.xlabel("Hour (0=Midnight, 23=11PM)")
plt.show()
```
